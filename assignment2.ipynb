```python
""" 7. Create a tuple:
data = (5, 10, 15, 20, 25, 30, 35)
Write a program to:
○ Count how many elements are divisible by 5
○ Find the sum of all elements
○ Find the average of the tuple
"""
data = (5, 10, 15, 20, 25, 30, 35)

count = 0
for i in data:
    if i%5==0:
        count += 1
print("count = " ,count)

sum = 0
for i in data:
    sum = sum + i
print("sum is :",sum)


average = sum/len(data)
print(average)
```

```python
""" 8. Create a dictionary:
students = {
"Aman": 78,
"Riya": 92,
"Kriti": 88,
"Rahul": 95
}
Write a program to:
○ Find the student with the highest marks
○ Find the student with the lowest marks
○ Print only the students who scored more than 85 marks
"""

students = {
    "Aman": 78,
    "Riya": 92,
    "Kriti": 88,
    "Rahul": 95
}

# Student with highest marks
highest_student = max(students, key=students.get)
print("Highest Marks:", highest_student, "->", students[highest_student])

# Student with lowest marks
lowest_student = min(students, key=students.get)
print("Lowest Marks:", lowest_student, "->", students[lowest_student])

# Students scoring more than 85 marks
print("Students scoring more than 85 marks:")
for name, marks in students.items():
    if marks > 85:
        print(name, "->", marks)
```

```python
""" 9. Write a function:
count_frequency(arr)
that takes a list as input and prints the frequency of each element.
Example:
arr = [1, 2, 2, 3, 1, 4, 2]
Output:
1 -> 2 times
2 -> 3 times
3 -> 1 time
4 -> 1 time
"""

arr = [1, 2, 2, 3, 1, 4, 2]
def count_frequency(arr):
    visited = []

    for item in arr:
        if item not in visited:
            count = 0

            for element in arr:
                if item == element:
                    count += 1

            if count == 1:
                print(f"{item} -> {count} time")
            else:
                print(f"{item} -> {count} times")

            visited.append(item)



count_frequency(arr)
```

```python
""" 10.Write a function:
find_duplicates(arr)
that takes a list as input and prints all duplicate elements present in
the list.
Example:
arr = [10, 20, 30, 20, 40, 10, 50, 30]
Output:
Duplicate elements are:
10
20
30
"""
def find_duplicates(arr):
    duplicates = []

    for item in arr:
        if arr.count(item) > 1 and item not in duplicates:
            duplicates.append(item)

    print("Duplicate elements are:")
    for item in duplicates:
        print(item)


arr = [10, 20, 30, 20, 40, 10, 50, 30]
find_duplicates(arr)
```
